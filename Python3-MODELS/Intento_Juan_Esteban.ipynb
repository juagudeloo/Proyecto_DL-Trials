```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20, 20)

class Read1:
    
    def __init__(self):
        
        self.ptm = ''
        self.filename = '154000'
        self.nx = 480
        self.ny = 480
        self.nz = 256
        
    
    def read(self,verbose=1):
       
        print("reading IOUT")
        self.iout = np.fromfile(self.ptm+"iout."+self.filename,dtype=np.float32)
        self.iout = np.reshape(self.iout, (self.nx, self.ny), order="C")
        print("IOUT done")    
        
        plot_iout = np.zeros(0)
        
        for i in range(self.nx):
                plot_iout = np.append(plot_iout, self.iout[i,:])
        
        plot_iout = np.reshape(plot_iout, (self.nx, self.ny))
        print(plot_iout)
        plt.matshow(plot_iout, cmap='gist_gray')
        plt.colorbar()
        plt.show()


if __name__ == "__main__":
    execute = Read1()
    execute.read()
```
